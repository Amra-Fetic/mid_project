1.What is the overall sales trend over time?

```python
# Exploring data
data.dtypes
```

```python
# Checking for missing values
data.isnull().sum()
```

```python
# Droping column with missing value
data = data.drop(columns=["Postal_Code"])
# Checking after drop
data.isnull().sum()
```

```python
# Checking for NaN's
data.isna().sum()
```

```python
# creating a plot so we can analyse sales over time 
import matplotlib.pyplot as plt
plt.figure(figsize=(20, 16))
plt.plot(data['Order_Date'], data['Sales'])
plt.title('Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.show()
```

The plot is very messy due to the granularity of data so it is neccessary to aggregate the date's by month so we can get smoother trend

```python
data['Order_Date'] = pd.to_datetime(data['Order_Date'], format='%d/%m/%Y')

# Extract month and year from the 'Order_Date' column to create a new column 'YearMonth'
data['YearMonth'] = data['Order_Date'].dt.to_period('M')

sales_column = 'Sales'
data_grouped = data.groupby('YearMonth')[sales_column].sum()

# Plot the data
plt.figure(figsize=(12, 6))
plt.plot(data_grouped.index.astype(str), data_grouped, label=sales_column, color = 'purple',marker='o')
plt.title('Monthly Grouped Data')
plt.xlabel('Year-Month')
plt.tight_layout()
# Rotate x-axis labels for better readability
plt.xticks(rotation=65)
plt.grid()
plt.legend()
plt.show()
```

2. Which shipping mode is most commonly used, and does it correlate with higher sales?

```python
# Loading library
import seaborn as sns
```

```python
# checking for unique values and counting how many times each ship mode is being used
shipping_mode_counts = data['Ship_Mode'].value_counts()
# Checking for most used ship mode 
most_common_shipping_mode = shipping_mode_counts.idxmax()
shipping_mode_counts_df= pd.DataFrame(shipping_mode_counts)
shipping_mode_counts_df
```

From this we can see that the most common ship mode is standard class

```python
display(most_common_shipping_mode)
```

```python
# visualization of previous claims 
plt.figure(figsize=(8, 6))
sns.countplot(x='Ship_Mode', data=data,palette='flare')
plt.title('Distribution of Shipping Modes')
plt.xlabel('Shipping Mode')
plt.ylabel('Count')
plt.show()
```

```python
# Calculating mean or average value for each group of ship mode 
avg_sales_by_mode = data.groupby('Ship_Mode')['Sales'].mean()
# Creating dataFrame for a better view
avg_sales_by_mode_df=pd.DataFrame(avg_sales_by_mode)
avg_sales_by_mode_df
```

```python
# Creating a barplot for visualization 
plt.figure(figsize=(10, 6))
sns.barplot(x='Ship_Mode', y='Sales', data=data, ci=None,palette='flare')
plt.title('Average Sales by Shipping Mode')
plt.xlabel('Shipping Mode')
plt.ylabel('Average Sales')
plt.show()
```

From this barplot we can conclude that there is no big difference between ship modes when it comes to the average sales

```python
descriptive_stats = data.groupby('Ship_Mode')['Sales'].describe()
display(descriptive_stats)
```

For the ship_mode 'First Class'sale values vary widely, with a high standard deviation. 'Same Day' - shipping is similar to 'First Class', the sales values show variability. 'Second Class'- sales values also have variability. 'Standard Class' - This shipping mode has the highest count of data points,sales values have a wide range, as indicated by the standard deviation.

```python
# loading the library
from scipy.stats import ttest_ind

#creating a boolean mask (True or False)to check in the columns ship mode whether the value is equal to 'Standard Class' or 'Same Day'

standard_sales = data['Sales'][data['Ship_Mode'] == 'Standard Class']
same_day_sales = data['Sales'][data['Ship_Mode'] == 'Same Day']
# conducting t stats and p value to make a decision about whether to reject the null hypothesis. 
t_stat, p_value = ttest_ind(standard_sales, same_day_sales)

print(f'T-test: t-statistic = {t_stat:.2f}, p-value = {p_value:.4f}')
```

In this case, a negative t-statistic suggests that the mean sales of 'Standard Class' is slightly lower than the mean sales of 'Same Day,' but the magnitude of the difference is small. With a p-value of 0.8929, we fail to reject the null hypothesis. There is not enough evidence to conclude that there is a significant difference in mean sales between 'Standard Class' and 'Same Day.' The difference observed in means could be due to random variability.

4. Is there a correlation between the segment of customers and the amount of sales?

First I wil do the sumamary of statistic for a different segments

```python
segment_sales_stats = data.groupby('Segment')['Sales'].describe()
segment_sales_stats_df = pd.DataFrame(segment_sales_stats)
segment_sales_stats_df
```

```python
# visualization of distribution for each segment 
plt.figure(figsize=(12, 6))
sns.boxplot(x='Segment', y='Sales', data=data)
plt.title('Distribution of Sales by Segment')
plt.xlabel('Segment')
plt.ylabel('Sales')
plt.show()
```

```python
# Check for unique values 
data['Segment'].unique()
```

```python
from scipy.stats import f_oneway

anova_result = f_oneway(
    data['Sales'][data['Segment'] == 'Consumer'],
    data['Sales'][data['Segment'] == 'Corporate'],
    data['Sales'][data['Segment'] == 'Home Office']
)

display(f'ANOVA F-statistic: {anova_result.statistic:.2f}')
display(f'P-value: {anova_result.pvalue:.4f}')
```

Since the p-value is higher than a common significance level, we do not have enough evidence to reject the null hypothesis.
The results suggest that there is no significant difference in the average sales values among the specified customer segments (Consumer, Corporatt, Home Office).

5. How do sales vary across different regions or countries?

```python
# Calculating descriptive statistics to understand the distribution of sales across different regions 
region_sales_stats = data.groupby('Region')['Sales'].describe()
region_sales_stats_df=pd.DataFrame(region_sales_stats)
region_sales_stats_df
```

East and South regions have higher mean sales compared to Central and West regions.
The South region has the highest standard deviation, indicating greater variability in sales.
The Central region has the lowest mean sales, while the South region has the highest mean sales.

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Region', y='Sales', data=data, ci=None)
plt.title('Average Sales by Region')
plt.xlabel('Region')
plt.ylabel('Average Sales')
plt.show()
```

```python
from scipy.stats import f_oneway

region_groups = [data['Sales'][data['Region'] == region] for region in data['Region'].unique()]
f_stat, p_value = f_oneway(*region_groups)

print(f'ANOVA F-statistic: {f_stat:.2f}')
print(f'P-value: {p_value:.4f}')
```

Based on the ANOVA results, we fail to reject the null hypothesis. This means that, statistically, there is not enough evidence to conclude that there are significant differences in mean sales values among the regions.
The variability in sales values observed among regions could be due to random chance rather than a true difference in population means.
